```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pop_df = pd.read_csv('Population.csv', skiprows=2, header=2)
pop_df.head()
```

```python
pop_df.info()
```

```python
pop_df.isna().sum()
```

```python
pop_df = pop_df.dropna()
```

```python
pop_df.isna().sum()
```

```python
pop_df.duplicated().sum()
```

```python
sns.displot(data=pop_df, x='2022', kind='hist',aspect=2, height = 8)
plt.title('Distribution of Population in Year 2022')
plt.xlabel('Year 2022')
plt.ylabel('Count')
plt.show()
```

```python
sns.catplot(data=pop_df, x='Country Name',y='2022',aspect=4, height=10, kind='bar')
plt.title('Country wise Population in 2022')
plt.xlabel('Country')
plt.ylabel('Year-2022')
plt.xticks(rotation=90)
plt.show()
```
